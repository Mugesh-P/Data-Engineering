```python
#Missing Values 
data.isnull().sum()
```

```python
data['price'] = data['price'].fillna(data['price'].median())
data['mileage'] = data['mileage'].fillna(data['mileage'].mean())
data['cylinders'] = data['cylinders'].fillna(method='ffill')
data['doors'] = data['doors'].fillna(data['doors'].median())
data.isnull().sum()
```

```python
# filling the NA values using KNNImputer

from sklearn.impute import KNNImputer
impute = KNNImputer()

for i in data.select_dtypes(include='number').columns:
    data[i] = impute.fit_transform(data[[i]])

data
```

```python
#standardizing dataset example

from sklearn.preprocessing import StandardScaler

df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5]
})

scaler = StandardScaler()

df['A'] = scaler.fit_transform(np.array(df['A']).reshape(-1, 1))

print(df)
```

```python
#outliers

data.boxplot(column='mileage')
```

```python
q3 = data['mileage'].quantile(0.75)
q1 = data['mileage'].quantile(0.25)

iqr = q3 - q1

lower_wisker = q1 - 1.5*iqr
upper_wisker = q3 + 1.5*iqr

data['mileage'] = np.where(data['mileage']<lower_wisker, lower_wisker, data['mileage'])
data['mileage'] = np.where(data['mileage']>upper_wisker, upper_wisker, data['mileage'])

data.boxplot(column='mileage')
```
